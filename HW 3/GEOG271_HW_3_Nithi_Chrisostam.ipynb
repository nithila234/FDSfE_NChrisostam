```python
#Loading USA state data
state_df = gpd.read_file("http://www2.census.gov/geo/tiger/GENZ2016/shp/cb_2016_us_state_5m.zip")
state_df.head()
```

#**Plotting Wisconsin's boundaries and data**

```python
import matplotlib.pyplot as plt
fig, gax = plt.subplots(figsize=(10, 10))
state_df.query("NAME == 'Wisconsin'").plot(ax=gax, edgecolor="black", color="white")
plt.show()
```

```python
#Load county data for Wisconsin state
county_df = gpd.read_file("http://www2.census.gov/geo/tiger/GENZ2016/shp/cb_2016_us_county_5m.zip")
county_df.head()
```

```python
county_df = county_df.query("STATEFP == '55'")
```

```python
#Plotting counties in Wisconsin
fig, gax = plt.subplots(figsize=(10, 10))
state_df.query("NAME == 'Wisconsin'").plot(ax=gax, edgecolor="black", color="white")
county_df.plot(ax=gax, edgecolor="black", color="white")
plt.show()
```

#**2016 Election Results in Wisconsin**

```python
results = pd.read_csv("https://datascience.quantecon.org/assets/data/ruhl_cleaned_results.csv", thousands=",")
results.head()
```

```python
results["county"] = results["county"].str.title()
results["county"] = results["county"].str.strip()
county_df["NAME"] = county_df["NAME"].str.title()
county_df["NAME"] = county_df["NAME"].str.strip()
```

```python
#Joining voting results with county geometry
res_states = county_df.merge(results, left_on="NAME", right_on="county", how="inner")
res_states.head()
```

```python
%%time
res_states["trump_share"] = res_states["trump"] / (res_states["total"])
res_states["rel_trump_share"] = res_states["trump"] / (res_states["trump"]+res_states["clinton"])
res_states.head()
```

```python
#Plotting a political map of the state of Wisconsin, with each county matched to voting data from 2016 elections
fig, gax = plt.subplots(figsize = (10,8))
state_df[state_df['NAME'] == 'Wisconsin'].plot(ax = gax, edgecolor='black',color='white')
res_states.plot(ax=gax, edgecolor='black', column='rel_trump_share', legend=True, cmap='RdBu_r', vmin=0.01, vmax=0.95)

#This map shows each Wisconsin county's share of votes towards Trump in the 2016 election
gax.annotate('Republican vote share',xy=(0.5, 0.05),  xycoords='figure fraction')

plt.axis('off')
plt.show()
```

```python
#How many counties did Trump win?
trump_counties = res_states.eval("trump > clinton").sum()
trump_counties
```

```python
#How many counties did Clinton win?
clinton_counties=res_states.eval("clinton > trump").sum()
clinton_counties
```

```python
#How many votes did Trump have in total?
trump_votes=res_states.eval("trump").sum()
trump_votes
```

```python
#How many votes did Clinton have in total?
clinton_votes=res_states.eval("clinton").sum()
clinton_votes
```

```python
#Who won?
trump_votes > clinton_votes
```

```python
#Vote difference between Trump and Clinton
trump_votes-clinton_votes
```
